## 1. Contexto de Negócio

A empresa 3A Superstore é uma rede varejista com múltiplas filiais e grande volume de transações. A empresa enfrenta dificuldades na retenção de clientes e deseja identificar padrões de comportamento que indiquem risco de churn.

## 2. Problema de Negócio

A empresa não possui um modelo preditivo capaz de identificar clientes com probabilidade de abandono, dificultando ações preventivas de retenção.

## 3. Objetivo do Projeto

Desenvolver um modelo preditivo para classificar clientes com risco de churn utilizando metodologia CRISP-DM.

## 4. Métrica de Sucesso

Melhorar identificação de clientes em risco

Aumentar eficiência de campanhas de retenção
