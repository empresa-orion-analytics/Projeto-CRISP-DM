```python
df_customers = spark.table("`crisp-dm`.customers")
df_orders = spark.table("`crisp-dm`.orders")
df_order_details = spark.table("`crisp-dm`.order_details")
df_branches = spark.table("`crisp-dm`.branches")
df_categories = spark.table("`crisp-dm`.categories")
```

```python
def clean_and_count(df, name):
    df_clean = df.dropna().dropDuplicates()
    
    total_original = df.count()
    total_clean = df_clean.count()
    
    print(f"===== {name} =====")
    print(f"Original: {total_original}")
    print(f"Limpo: {total_clean}")
    print(f"Removidos: {total_original - total_clean}")
    print()
    
    return df_clean
```

```python
df_customers_clean = clean_and_count(df_orders, "Customers")
df_orders_clean = clean_and_count(df_customers, "Orders")
df_order_details_clean = clean_and_count(df_order_details, "Order Details")
df_branches_clean = clean_and_count(df_branches, "Branches")
df_categories_clean = clean_and_count(df_categories, "Categories")
```
